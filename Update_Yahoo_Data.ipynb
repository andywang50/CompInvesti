```python
import pandas_datareader as pdr
from datetime import datetime
import QSTK.qstkutil.DataAccess as da
import urllib2
import urllib

c_dataobj = da.DataAccess('Yahoo')
ls_all_syms = c_dataobj.get_all_symbols()
#ls_all_syms =  ["AAPL", "GLD", "GOOG", "$SPX", "XOM"]
data_path = '/Users/andy/anaconda/envs/py27/lib/python2.7/site-packages/QSTK/QSData/Yahoo/'

miss_ctr = 0
ls_missed_syms = []


for symbol in ls_all_syms:
    symbol_name = symbol
    if symbol == "$SPX":
        symbol = "SPY"
    if symbol[0] == '$':
        symbol = '^' + symbol[1:]
    try:
        df = pdr.get_data_yahoo(symbols=symbol, start=datetime(2000, 1, 1), end=datetime(2018, 4, 18))
        df = df.iloc[::-1]
        ## Note: the order of the column MUST be as this!
        df = df[['Open','High','Low','Close','Volume','Adj Close']]
        df.to_csv(data_path + symbol_name+ '.csv',sep=',',float_format='%.2f') ## keep two decimals
    
    except urllib2.HTTPError:
        miss_ctr += 1
        ls_missed_syms.append(symbol_name)
        print "Unable to fetch data for stock: {0}".format(symbol_name)
    except urllib2.URLError:
        miss_ctr += 1
        ls_missed_syms.append(symbol_name)
        print "URL Error for stock: {0}".format(symbol_name)
    except:
        miss_ctr += 1
        ls_missed_syms.append(symbol_name)
        print "Unable to fetch data for stock: {0} for unknown error".format(symbol_name)
print "All done. Got {0} stocks. Could not get {1}".format(len(ls_all_syms) - miss_ctr, miss_ctr)
print ls_missed_syms
# ibm = pdr.get_data_yahoo(symbols='^VIX', start=datetime(2000, 1, 1), end=datetime(2018, 4, 18))
# ibm = ibm.iloc[::-1]
# ibm.to_csv('IBM.csv')
# ibm
```
