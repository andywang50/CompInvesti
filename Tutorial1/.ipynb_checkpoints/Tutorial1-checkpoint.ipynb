```python
import numpy as np
import QSTK.qstkutil.qsdateutil as du
import QSTK.qstkutil.tsutil as tsu
import QSTK.qstkutil.DataAccess as da

import datetime as dt
import matplotlib.pyplot as plt
import pandas as pd

ls_symbols = ["AAPL", "GLD", "GOOG", "$SPX", "XOM"]
#ls_symbols = ["AAPL", "$SPX"]
dt_start = dt.datetime(2018, 4, 1)
dt_end = dt.datetime(2018, 4, 18)
dt_timeofday = dt.timedelta(hours=16) # time duration
ldt_timestamps = du.getNYSEdays(dt_start, dt_end, dt_timeofday)  # the days that NYSE is open

c_dataobj = da.DataAccess('Yahoo')
ls_keys = ['open', 'high', 'low', 'close', 'volume', 'actual_close']
ldf_data = c_dataobj.get_data(ldt_timestamps, ls_symbols, ls_keys)
d_data = dict(zip(ls_keys, ldf_data))
```

```python

na_price = d_data['close'].values
datetimes = [d.to_pydatetime() for d in ldt_timestamps]
plt.clf()
f, ax = plt.subplots()
ax.set_title('close prices')
ax.set_ylabel('Close')
ax.set_xlabel('Date')
ax.plot(datetimes, na_price)
ax.legend(ls_symbols,loc='best')
f.savefig('close.pdf', format='pdf')
plt.show()

```

```python
na_normalized_price = na_price / na_price[0, :]
plt.clf()
f,ax=plt.subplots()
ax.set_title('normalized close prices')
ax.set_ylabel('normalized close price')
ax.set_xlabel('Date')
ax.plot(datetimes, na_normalized_price)
ax.legend(ls_symbols,loc='best')
f.savefig('normalized close.pdf', format='pdf')
plt.show()
```
